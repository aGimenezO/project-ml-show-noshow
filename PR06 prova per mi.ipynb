```python
#Convertir variables categóricas en numéricas (0 y 1)

#No = 0       Yes = 1
  
#Male = 0     Female = 1
```

```python
#First method - using replace:

gender_encode = {"Gender":{"M": 0, "F": 1}}

df.replace(gender_encode, inplace=True)

df.head()
```

```python
noshow_encode = {"No-show":{"No": 0, "Yes": 1}}

df.replace(noshow_encode, inplace=True)

df.head()
```

```python
#Me reconoce a Gender como Float y a No-Show como Integer

df.info()
```

```python
df["Gender"] = df["Gender"].astype('int64')
df.dtypes
```

```python
#Second method - Using scikit learn

from sklearn.preprocessing import LabelEncoder
```

```python
lb_make = LabelEncoder()
```

```python
df.head()
```

```python
df["Gender_code"] = lb_make.fit_transform(df["Gender"])
```
