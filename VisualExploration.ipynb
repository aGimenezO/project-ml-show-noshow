```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = pd.read_csv('data/show_no_show_ready_to_model.csv')
```

```python
df.head()
```

## Age Distribution

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.style.use('seaborn')

df["Age"].plot(kind='hist', rot=0, bins = 200)
plt.xlabel('Age')
plt.ylabel('Quantity')
plt.title('Distribution of Age in the data set')

plt.show()
```

## Gender Quantity

```python
Gender_M=df.Gender_M.value_counts()
Gender_M #F = 0, M=1
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.style.use('seaborn')

Gender_M.plot(kind="bar")
plt.xlabel('Gender')
plt.ylabel('Quantity')
plt.title('Quantity of female and male')
ticks = ["Female (71504)","Male (38514)"]
ax.set_xticklabels(ticks)

plt.show()
```

## No show quantity

```python
No_show=df["No-show"].value_counts()
No_show
```

```python
fig, ax = plt.subplots(1, figsize=(8,6))
plt.style.use('seaborn')

No_show.plot(kind="bar")
plt.xlabel('No Show - Show')
plt.ylabel('Quantity')
plt.title('Quantity of Show and No Show')
ticks = ["No Show (87823)","Show (22195)"]
ax.set_xticklabels(ticks)

plt.show();
```

## Quantity of Show and No-Show per Gender

```python
sns.set(style="whitegrid")

g = sns.countplot(df["No-show"], hue=df.Gender_M)
#g.set_yscale("log")
# No-show(No = 0, Yes=1)
# Gender( female= 0 and male= 1)
```

## Quantity of Gender per Show and No-Show 

```python
ax = sns.countplot(x=df.Gender_M, hue=df["No-show"])
ax.set_title("Show/NoShow per Females and Males")
x_ticks_labels=['Female', 'Male']
ax.set_xticklabels(x_ticks_labels)
plt.show()
```
